```python
from os import walk
import numpy as np
import cv2
import scipy as scipy
from scipy.fft import fft, ifft
import matplotlib.pyplot as plt
from astropy.io import fits
import polarTransform
from astropy.utils.data import get_pkg_data_filename
from scipy.stats import wasserstein_distance

from marvin.tools import Maps

from tqdm import tqdm
import Bicone_Classifier
```

## Iterate through the galaxies

```python
HEADER_PATH = '/Users/runquanguan/Documents/bicone_candidates_v2.fits'
hdul = fits.open(HEADER_PATH)
hdu = hdul[1].data
plateifu = hdu['PLATEIFU']

result_list = []

for data in tqdm(plateifu):
    try:
        plateifu = data
        BC_EW_hist = Bicone_Classifier.generate_profile_histogram(plateifu)
        yf = np.abs(fft(BC_EW_hist-0.5)[1:50])
        if yf[1] <= 60 and yf[5] <= 45:
            result_list.append(data)
        else:
            pass
    except:
        pass
```

```python
HEADER_PATH = '/Users/runquanguan/Documents/bicone_candidates_v3.fits'
hdul = fits.open(HEADER_PATH)
hdu = hdul[1].data
plateifu = hdu['PLATEIFU']

result_list = []
miss_list = []

for data in tqdm(plateifu):
    # Load OIII MAPS from SDSS-MARVIN server
    try:
        curve = Bicone_Classifier.ellip_ring_ew(data, in_r = 1.1, out_r = 1.3, cycle=2)[0]
        fourier = Bicone_Classifier.fourier_classifier(curve)
        

        if fourier[0] == 3 and fourier[1] >= 30:
            result_list.append(data)
    except:
        miss_list.append(data)


```

```python
c1 = fits.Column(name='plateifu', array=np.array(result_list), format='20A')
t = fits.BinTableHDU.from_columns([c1])
t.writeto('bicone_candidates_v4.fits')
```
